```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
S =iris.data
t =iris.target
target =iris.target_names
```

```python
from sklearn.preprocessing import StandardScaler
sclr =StandardScaler()
S_val = sclr.fit_transform(S)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda =LDA(n_components=2)
S_lda =lda.fit_transform(S_val,t)
```

```python
lda_df =pd.DataFrame(data=S_lda,columns=['LD1','LD2'])
lda_df['target'] =t
plt.figure(figsize=(12,8))
sns.scatterplot(data=lda_df,x='LD1',y='LD2',hue='target',palette='Set2',s=100)
plt.title('LDA of Iris Dataset')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend(title='Classes',labels=target)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca =PCA(n_components=2)
S_pca =pca.fit_transform(S_val)
PCA_df =pd.DataFrame(data=S_pca,columns=['PC1','PC2'])
PCA_df['target'] =t
plt.figure(figsize=(12,8))
sns.scatterplot(data=PCA_df,x='PC1',y='PC2',hue='target',palette='Set2',s=100)
plt.title('PCA of Iris Dataset')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Classes',labels=target)
plt.show()
```

```python
fig, axes =plt.subplots(1,2,figsize=(16,6))
sns.scatterplot(data=lda_df,x='LD1',y='LD2',hue='target',palette='Set2',s=100,ax=axes[0])
axes[0].set_title('LDA of Iris Dataset')
axes[0].set_xlabel('LD1')
axes[0].set_ylabel('LD2')
axes[0].legend(title='Classes',labels=target)
sns.scatterplot(data=PCA_df,x='PC1',y='PC2',hue='target',palette='Set2',s=100,ax=axes[1])
axes[1].set_title('PCA of Iris Dataset')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].legend(title='Classes', labels=target)

plt.show()
```
