```python
import pandas as pd
ds = pd.read_csv('/content/enjoysports.csv')
ds.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

M=ds.drop('enjoysport',axis=1)
n=ds['enjoysport']
M=pd.get_dummies(M,drop_first=True)

M_train,M_test,n_train,n_test=train_test_split(M,n,test_size=0.2,random_state=42)

model=GaussianNB()
model.fit(M_train,n_train)

n_pred=model.predict(M_test)

con_Matrix=confusion_matrix(n_test,n_pred)
print("Confusion Matrix computed is:\n",con_Matrix)

Accuracy=accuracy_score(n_test,n_pred)
print("Accuracy value is:",Accuracy)

Error_Rate=1-Accuracy
print("Error Rate value is:",Error_Rate)

class_rep=classification_report(n_test,n_pred)
print("Classification Report(Precision, Recall, F1-score, Support) is:\n",class_rep)
```
